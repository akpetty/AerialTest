# Random Forest
- General concept of using multiple decision trees to convert various data streams that appear to be “weak learners” to collectively become a “strong learner”
- Use the scikit-learn Random Forest alogirthm
- Need to clean up the data first and generate training/test data.
- Did originally use the different years as Train/Test, but I've decided it best to combine years and subset using a fraction of the total data

```python
# Combine the two years of data
dfAll = pd.concat([df1, df2])
dfTrain=dfAll.sample(frac=0.8,random_state=200)
dfTest=dfAll.drop(dfTrain.index)
#dfTrain.shape, dfTest.shape

# Get training/test yield datasets
yieldTrain=dfTrain['Yield']
yieldTest=dfTest['Yield']
# Get lon/Lat of the test yields to plot out later
lonT=dfTest['Longitude'][dfTest.index]
latT=dfTest['Latitude'][dfTest.index]

# Drop the yield column from the fitting datasets
dfTrain=dfTrain.drop('Yield',axis=1)
dfTest=dfTest.drop('Yield',axis=1)

# Drop Lat/Lon to make it a harder (fairer?) test..
#dfTrain=dfTrain.drop('Longitude',axis=1)
#dfTrain=dfTrain.drop('Latitude',axis=1)
#dfTest=dfTest.drop('Longitude',axis=1)
#dfTest=dfTest.drop('Latitude',axis=1)
dfTrain.shape, dfTest.shape 
```

```python
def scale_all_features(dfT):  
    ''' Function to normalize all columns in a Pandas dataFrame
    '''
    features = list(dfT.columns)
    dfT[features] = dfT[features].apply(lambda x: (x-x.min())/(x.max()-x.min()), axis=0)  
    return dfT
```

```python
# Normalize data (note that Yields aren't normalized as no reason to)
dfTest = scale_all_features(dfTest)
dfTrain = scale_all_features(dfTrain)
dfTest.head(2)
dfTrain.head(2)
```

```python
# Do something with the NaN values. Should probably think about this more.
features = list(dfTrain.columns)
for x in features:
    # Fill the nan values with the median value, or fitting model isn't happy.
    dfTrain[x].fillna(dfTrain[x].median(), inplace=True)
    dfTest[x].fillna(dfTest[x].median(), inplace=True)
```

```python
# Yields need to be integers (or strings) to be used in the random forest classifier
yieldTrainS = np.asarray(yieldTrain, dtype="int")
yieldTestS = np.asarray(yieldTest, dtype="int")
#yieldTrainS = np.asarray(yieldTrain, dtype="|S6")
#yieldTestS = np.asarray(yieldTest, dtype="|S6")
yieldTestS
```

```python
def rms(var):
    ''' return the RMSE of 1D array
    '''
    return sqrt(sum([x**2 for x in var])/size(var))
```

```python
# Here we scan through a chosen parameter space to find the best Random Forest framework model to use. 
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import StratifiedKFold
from sklearn.grid_search import GridSearchCV
from scipy import stats

forest = RandomForestClassifier()
cross_validation = StratifiedKFold(yieldTrainS, n_folds=5)
# just explore the number of estimators and forest index
parameter_grid = {   
                 'n_estimators': [60, 80, 100],
                 'criterion': ['gini', 'entropy']
                 }
grid_search = GridSearchCV(forest,param_grid=parameter_grid)
                           #cv=cross_validation)

# Alternatively, if we know what parameters we want to set, we can just comment out the above and use the following
#forest = RandomForestClassifier(n_estimators = 19, n_jobs=-1)
#model = forest.fit(dfTrain,yieldTrainS)
#guess = model.predict(dfTest)

grid_search.fit(dfTrain, yieldTrainS)
print('Best score: {}'.format(grid_search.best_score_))
print('Best parameters: {}'.format(grid_search.best_params_))

guess = grid_search.predict(dfTest)
slope, intercept, r, prob, stderr = stats.linregress(guess, yieldTestS)
rmse=rms(guess-yieldTestS)
r, rmse
# NB My computer struggled to process the mdoel with n_estimators much higher than 100 which was a limiting factor...
```

```python
# Look at the scatter plot of guess and predictions
fig =figure(figsize=(8,8))
scatter(guess, yieldTestS)
plot(np.linspace(10,70,10), intercept+ slope*np.linspace(10,70,10))
xlabel('Yield')
ylabel('Yield')
```

```python
# Plot data for the test data and predicted
#generate map data from those lon/lats genereated earlier
xptsT, yptsT = m(np.array(lonT), np.array(latT))

fig =figure(figsize=(14,7))
ax1=subplot(121)
im1 = m.scatter(xptsT, yptsT, c=yieldTestS,cmap=cm.cubehelix_r, vmin=0, vmax=80, zorder=2)
m.drawparallels(np.arange(25,65,20))
m.drawmeridians(np.arange(-120,-40,20))
m.drawcoastlines()
#m.fillcontinents()
ax1.annotate('Test data', xy=(0.02, 1.02), fontsize=10,xycoords='axes fraction')
ax2=subplot(122)
im2 = m.scatter(xptsT, yptsT, c=guess,cmap=cm.cubehelix_r, vmin=0, vmax=80,zorder=2)
m.drawparallels(np.arange(25,65,20))
m.drawmeridians(np.arange(-120,-40,20))
m.drawcoastlines()
#m.fillcontinents()
ax2.annotate('Prediction', xy=(0.02, 1.02), fontsize=10, xycoords='axes fraction')
cax1 = fig.add_axes([0.92, 0.3, 0.03, 0.4])
cbar = colorbar(im2,cax=cax1, orientation='vertical', extend='both', use_gridspec=True)
cbar.set_label('Yield')
plt.show()
```

# Final thoughts
- The random forest appears to work very well for predicting crop yields (based on correlation and RMSE), although it's tough to tell what might be viewed as a good/skillful prediction in this case.
- The location (lon/lat) are key variables, which obviously makes sense. They likely integrate several factors related to the loca environment not included in the atmospheric variables. Removing them from the model caused a very big drop in predictive skill, which may be an issue if you didn't want location to come into it.     
- I would have liked to have explored alternative machine learning techniques if I had more time (e.g. a neural network or naive Bayes), but my experience sugegsted random forest would be pretty likely to work reasonably well at the very least.
